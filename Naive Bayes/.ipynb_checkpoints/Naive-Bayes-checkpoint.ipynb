```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('credit card.csv')
df.head()
```

```python
df.shape
```

# **Missing Value Checking & Replacing with Mean**

```python
df.isnull().sum()
```

```python
age_mean = df['AGE'].mean()
age_mean = round(age_mean, 2)
print(age_mean)
```

```python
df['AGE'] = df['AGE'].fillna(age_mean)
```

```python
df.head()
```

# **Feature & Target Choose for classification**

```python
X = df.drop('default.payment.next.month', axis = 1)
Y = df['default.payment.next.month']
```

# **Train Test Split**

```python
from sklearn.model_selection import train_test_split
xtrain, xtest, ytrain, ytest = train_test_split(X, Y, test_size = 0.25, random_state = 42)
```

```python
xtrain
```

# **Naive Bayes Classifier Model**
## **Gausian Naive Bayes**

```python
from sklearn.naive_bayes import GaussianNB

GNB = GaussianNB()

GNB.fit(xtrain, ytrain)
```

```python
GNB.score(xtest, ytest)
```

## **Bernouli Naive Bayes**

```python
from sklearn.naive_bayes import BernoulliNB

BNB = BernoulliNB()

BNB.fit(xtrain, ytrain)
```

```python
BNB.score(xtest, ytest)
```

## **Multinomial Naive Bayes**

```python
from sklearn.naive_bayes import MultinomialNB

MNB = MultinomialNB()

MNB.fit(xtrain, ytrain)
```

<h2 style='color:Blue'><b>So, Bernouli Naive Bayes is performing well in these data with Accuracy of 76.8% and <span style='color:Red'>Multinomial NB can't pass the Negative value containing dataset</span></b></h2>
